```python
def getDf(url) :
    api_key = '546a726a616a683039336b46467648'
    urlRe = url.replace('(인증키)',api_key).replace('xml','json').replace('/5/','/1000/')
    res = requests.get(urlRe).json()
    df = pd.DataFrame(res[url.split('/')[-4]]['row'])
    return df
```

```python
df = getDf(url)
```

```python
sel_c = ['SGG_NM','BJDONG_NM','LAND_GBN_NM','FLR_NO','CNTRCT_DE','RENT_GBN','RENT_AREA','RENT_GTN','RENT_FEE','BLDG_NM','BUILD_YEAR']
```

```python
myDf = df[sel_c]
```

```python
myDf.columns=['자치구명','법정동명','지번구분명','층수','계약일','전월세','면적','보증금','임대료','건물명','건축년도']
```

```python
newDf = myDf[myDf.전월세 == '전세']
```

```python
newDf.astype({'보증금':'int','임대료':'int'})
```

```python
ddf = newDf.groupby('자치구명')[['보증금','임대료']].mean().round()
```

```python
ddf.sort_values(by=['임대료'],ascending=False)
```

```python
ddf['x'] = round(ddf.임대료/ddf.보증금,4)
```

```python
ddf.sort_values(by=['x'],ascending=False)
```
